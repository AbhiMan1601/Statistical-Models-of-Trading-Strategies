```python
file = pd.read_csv("/kaggle/input/average/HINDALCO_1D.xlsx - HINDALCO.csv")
file.head()
```

```python
file.info()
```

```python
file.drop('Unnamed: 7', inplace=True, axis=1)
```

```python
file.head()
```

```python
# Sum of return
data = file['close']
file['log_return'] = np.log(data/data.shift())
print(file['log_return'])
```

```python
file.head()
```

```python
file1 = file.drop(0)
file1.head()
```

```python
import seaborn as sns
%matplotlib inline
sns.histplot(data=file1, y="log_return")

# we see that the returns are centred around 0 : less risky/volatile
# log returns are normally distributed
```

```python
sharpe_ratio = (0-file1['log_return'].mean())/file1['log_return'].std()
print(sharpe_ratio)
# This is the daily Sharpe ratio
```

```python
import math

math.exp(-0.022059848105174827)
# Since sharpe ratio in the normal scale _=1, we say that this is a good enough value for investment at the given level of log return
```

```python
def information_ratio(returns, benchmark_returns, days=252):
 return_difference = returns - benchmark_returns
 volatility = return_difference.std() * np.sqrt(days)
 information_ratio = return_difference.mean() / volatility
 return information_ratio

information_ratio(0,file1['log_return'])
```
